```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

%matplotlib inline
```

```python
l256_runtimes_file = "experiments/Lublin_256_runtimes.csv"
l256_estimate_file = "experiments/estLublin_256_estimate.csv"
l1024_runtimes_file = "experiments/Lublin_1024_runtimes.csv"
l1024_estimate_file = "experiments/estLublin_256_estimate.csv"
```

```python
l256_runtimes_data = pd.read_csv(l256_runtimes_file)
l256_estimate_data = pd.read_csv(l256_estimate_file)
l1024_runtimes_data = pd.read_csv(l1024_runtimes_file)
l1024_estimate_data = pd.read_csv(l1024_estimate_file)
```

```python
l256_runtimes_data = pd.melt(l256_runtimes_data)
l256_runtimes_data.rename(columns={'variable': 'Scheduling policies','value':'Average bounded slowdown'}, inplace=True)

l256_estimate_data = pd.melt(l256_estimate_data)
l256_estimate_data.rename(columns={'variable': 'Scheduling policies','value':'Average bounded slowdown'}, inplace=True)

l1024_runtimes_data = pd.melt(l1024_runtimes_data)
l1024_runtimes_data.rename(columns={'variable': 'Scheduling policies','value':'Average bounded slowdown'}, inplace=True)

l1024_estimate_data = pd.melt(l1024_estimate_data)
l1024_estimate_data.rename(columns={'variable': 'Scheduling policies','value':'Average bounded slowdown'}, inplace=True)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14,8))

fig.suptitle('Synthetic workload trace performance experiments')

axes[0,0].set_title('Using the actual processing time with 256 processors')
sns.boxplot(ax=axes[0,0], x="Scheduling policies", y="Average bounded slowdown", data=l256_runtimes_data, linewidth=0.5, palette="Pastel1", fliersize=0)
sns.stripplot(ax=axes[0,0], x="Scheduling policies", y="Average bounded slowdown", hue="Scheduling policies", data=l256_runtimes_data, linewidth=0.5, size=3, jitter=True, palette="Pastel1")
axes[0,0].get_legend().remove()

axes[0,1].set_title('Using the estimate processing time with 256 processors')
sns.boxplot(ax=axes[0,1], x="Scheduling policies", y="Average bounded slowdown", data=l256_estimate_data, linewidth=0.5, palette="Pastel1", fliersize=0)
sns.stripplot(ax=axes[0,1], x="Scheduling policies", y="Average bounded slowdown", hue="Scheduling policies", data=l256_estimate_data, linewidth=0.5, size=3, jitter=True, palette="Pastel1")
axes[0,1].get_legend().remove()

axes[1,0].set_title('Using the actual processing time with 1024 processors')
sns.boxplot(ax=axes[1,0], x="Scheduling policies", y="Average bounded slowdown", data=l1024_runtimes_data, linewidth=0.5, palette="Pastel1", fliersize=0)
sns.stripplot(ax=axes[1,0], x="Scheduling policies", y="Average bounded slowdown", hue="Scheduling policies", data=l1024_runtimes_data, linewidth=0.5, size=3, jitter=True, palette="Pastel1")
axes[1,0].get_legend().remove()

axes[1,1].set_title('Using the estimate processing time with 1024 processors')
sns.boxplot(ax=axes[1,1], x="Scheduling policies", y="Average bounded slowdown", data=l1024_estimate_data, linewidth=0.5, palette="Pastel1", fliersize=0)
sns.stripplot(ax=axes[1,1], x="Scheduling policies", y="Average bounded slowdown", hue="Scheduling policies", data=l1024_estimate_data, linewidth=0.5, size=3, jitter=True, palette="Pastel1")
axes[1,1].get_legend().remove()

plt.tight_layout()
```
