```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
sns.set_style("whitegrid")
%matplotlib inline
```

```python
columns_label=['p', 'q', 'r', 'score']

sim0 = pd.read_csv("score-distribution-0.csv", header=None, names=columns_label).assign(Simulation=0)
sim1 = pd.read_csv("score-distribution-1.csv", header=None, names=columns_label).assign(Simulation=1)
```

```python
# convert to float
for colname in sim1[['p','q','r']]:
    sim1[colname] = sim1[colname].astype('float')
sim1.dtypes
```

```python
concat_sims = pd.concat([sim0, sim1])    
```

```python
display(concat_sims.head())
```

```python
display(concat_sims.info())
```

# Informative plots

```python
fig, axes = plt.subplots(2, 2, figsize=(8,7))
fig.suptitle(r'Distribuition of $p$, $q$, $r$ and $score$.')

sns.boxplot(ax=axes[0,0], data=concat_sims, x='Simulation', y='p', orient='v', showfliers=False, width=0.3, palette="Set2")
sns.boxplot(ax=axes[0,1], data=concat_sims, x='Simulation', y='q', orient='v', showfliers=False, width=0.3, palette="Set2")
# sns.boxplot(ax=axes[1,0], data=concat_sims, x='Simulation', y='r', orient='v', showfliers=False, width=0.3)
sns.histplot(ax=axes[1,0], data=concat_sims, x='r', hue='Simulation', kde=True, palette="Set2")
sns.boxplot(ax=axes[1,1], data=concat_sims, x='Simulation', y='score', orient='v', showfliers=False, width=0.3, palette="Set2")

plt.tight_layout()
```

Both simulations seem equivalent. Merging them would change a bit, nothing compromising.
